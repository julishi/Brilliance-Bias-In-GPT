```python
import seaborn
import pandas
import matplotlib.pyplot as plt
f_path = "/content/gdrive/MyDrive/SCU_Grad/Research/Dr. Ackerman/GPT-3-Brilliance-Bias/InstructGPT-Traits-Men-Women - Women.csv"
csv = pandas.read_csv(f_path, names=['word', 'correlation degree'])
plt.figure(figsize = (80,35))
sns.set_style("whitegrid")

ax = sns.barplot(x = 'word', y = 'correlation degree',data =csv, palette=sns.color_palette("rocket"))#.set(title='Highest 30 Adjectives & Verbs')
ax.tick_params(labelsize=35)
ax.set_xlabel('Word', fontsize=50)
ax.set_ylabel('Correlation Degree', fontsize=50)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 45, fontsize=50)
ax.set_title("InstructGPT Female Highest 30 Correlating Adjectives & Verbs", fontsize=60)
plt.savefig('InstructGPT-top-words-women.png')
```

```python
import seaborn
import pandas
import matplotlib.pyplot as plt
f_path = "/content/gdrive/MyDrive/SCU_Grad/Research/Dr. Ackerman/GPT-3-Brilliance-Bias/InstructGPT-Traits-Men-Women - Lexicons.csv"
csv = pandas.read_csv(f_path, header=0)
csv['Type'] = csv['Type'].str.upper()
plt.figure(figsize = (50,25))
sns.set(font_scale = 1.5)
sns.set_style("ticks")
fig, ax1 = plt.subplots(figsize=(10, 10))
tidy = csv.melt(id_vars='Type').rename(columns=str.title)
ax = seaborn.barplot(x='Type', y='Value', hue='Variable', data=tidy, ax=ax1, palette=sns.color_palette("rocket"))
#plt.xlabel('Type', fontsize=15)
#plt.ylabel('Value', fontsize=15)
ax.set_title("InstructGPT Lexicon Analysis")
plt.savefig('InstructGPT-Lexicon-Analysis.png')
```
